```sql
SELECT
  model_key,
  model_name,
  model_version,
  prediction,
  COUNT(*) AS total_predictions,
  AVG(score) AS avg_score
FROM workspace.financial_sentiment.gold_financial_sentiment_predictions
GROUP BY
  model_key,
  model_name,
  model_version,
  prediction
ORDER BY
  model_key,
  prediction;
```
