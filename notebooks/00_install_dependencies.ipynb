# 00 Install Dependencies

This notebook installs the project dependencies from `requirements.txt`.

Important: the TensorFlow dependency in this project requires Python `3.10` to `3.13`. If you run this notebook with Python `3.14`, installation will fail before TensorFlow is installed.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

requirements_path = ROOT / "requirements.txt"
python_version = sys.version_info[:2]

print(f"Project root: {ROOT}")
print(f"Requirements: {requirements_path}")
print(f"Python: {sys.version.split()[0]}")

if not requirements_path.exists():
    raise FileNotFoundError(f"Could not find requirements file: {requirements_path}")

if python_version < (3, 10) or python_version > (3, 13):
    raise RuntimeError(
        "This project's TensorFlow dependency requires Python 3.10 to 3.13. "
        f"Current version: {sys.version.split()[0]}"
    )
```
